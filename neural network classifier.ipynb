```python
import numpy as np
import pandas as pd
#tensor flow dep learning ma use huti ha
import tensorflow as tf
```

```python
tf.__version__
```

```python
data=pd.read_csv('Churn_Modelling.csv')
```

```python
data.head()
```

```python
#humne sirf pehla 3 columns chorna ha aur akrhi q k yrh hmara features ha so
x=data.iloc[:,3:-1].values

#label sirf akhri column ha
y=data.iloc[:,-1].values
```

```python
x.shape
```

```python
y.shape
```

```python
# jisma 2 option wala categorical data  encode karna hu waha lagata ha jisma order matter karta ha
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
# column 2 ko encode karege q k wo categorical data ha hmara male female likha ha jisma
x[:,2]=le.fit_transform(x[:,2])
```

```python
# country wala jo column reh gya ha k jis ma order matter nhi karta usma hum OneHotEnocding karega q k options zyada ha isma
from sklearn.compose import ColumnTransformer 
from sklearn.preprocessing import OneHotEncoder

'''yha ColumnTransformer sara column ki values badlega usma jo prameter diya ha 'encoder' uska matlab ha encoding karni ha uska btaya ha OneHotEncoding karni column [1] pe
# remainder means baqi columns ko chorna ha
sara kuch hu ker k value transformers ko milegi aur wo assign hugi ct ko
'''
ct=ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[1])],remainder='passthrough')
x=np.array(ct.fit_transform(x))

#france germany sare country encode ho gai ha
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
#deep learning ma hama har hal ma feature scaling apply karni ha chaha data same scale per lie kare ya na kare
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
x_train.shape
```

```python
# create artificial neural network 
#yha keras API ha aur tf tensor flow ka object ha
ann= tf.keras.models.Sequential()

# yeh 1st hidden layer ha
#units ki value problem to problem vary kar sakti ha aur activation function is k ilawa prelu ha, sigmoid ha 
#wo dekh kar use kar la
ann.add(tf.keras.layers.Dense(units=6,activation='relu'))

#yeh 2nd hidden layer ha
ann.add(tf.keras.layers.Dense(units=6,activation='relu'))

#yeh 2 neurons bnai ha
```

```python
# outptut layer design karni ha
#ager tu binary number k liye probability nikal rha ha tu unit=1 ok ha
'''mager ap multi features classification kar rha ha jaisi country detect karna ha ya pics ha 
tu number of categories k equal unit ko rakhna ha'''
#binary class k liye function sigmoid use hota ha mager multi class classification k liye koi aur dekh sakte ha

ann.add(tf.keras.layers.Dense(units=1,activation='sigmoid'))  #sofnix multi class classification k liya sigmoid ki jga
```

```python
#training karni ha isma neural network ki
'''yha loss function ha jo fraq nikalta ha actual value aur predicted value k dermyan
phr usko kam karta ha her iteration k sath
# loss=binary_cross_entropy  ha binary classification k liye 
ager multi class classification ki huti tu categorical_cross_entropy'''

ann.compile(optimizer='adam', loss='binary_crossentropy',  metrics=['accuracy'])

#yha batch size =32 ka matlab ha k her batch k loss funtion nikla ga aur epochs ko machine power dekh k barhana ha 1000 tak leja sakte hu
ann.fit(x_train,y_train, batch_size =32, epochs=100)
```

```python
#uper hum na x_test aur x_train dono ko transform k sath diya tha tu idhr bhi transform karege
#country france ha 1,0,0 phr 
print(ann.predict(sc.transform([[1,0,0,600,1,40,3,60000,2,1,1,50000]])) > 0.5)
```

```python
y_pred=ann.predict(x_test)
y_pred=(y_pred > 0.5)                               # y_pred ka answer 0.5 se zyada ayega tu value 1 hu other wise 0 

#end wala 1 columns ki shakal ma dono ka result likha ga
print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
acc=accuracy_score(y_test,y_pred)
cm=confusion_matrix(y_test,y_pred)
print(cm)
print(acc)
```
