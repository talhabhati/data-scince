```python
import seaborn as sns
%matplotlib inline
```

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
#lmplot plot graph dega jis k sath aik best fit line bhi hogi
sns.lmplot(x='total_bill',y='tip',data=tips)
```

```python
#smoker aur non smoker k liye alag alag bestfit line ayegi
sns.lmplot(x='total_bill',y='tip',data=tips,hue='smoker')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='smoker',markers=['o','v'],scatter_kws={'s':100})
```

```python
#col='sex'  yeh alag alag grap
sns.lmplot(x='total_bill',y='tip',data=tips,hue='smoker',markers=['o','v'],scatter_kws={'s':100},col='sex')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,scatter_kws={'s':100},col='sex')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,scatter_kws={'s':100},col='sex',row='time')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',col='day',aspect=0.8,size=8)
```
